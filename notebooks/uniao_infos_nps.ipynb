```python
import pandas as pd
```

```python
df = pd.read_csv('../dados/dados_totvs/nps_relacional.csv', sep=';')
```

```python
df
```

```python
df = df.drop(['respondedAt'], axis=1)
```

```python
df.isnull().sum()/len(df)
```

```python
colunas_para_preencher = df.columns.tolist()
```

```python
colunas_para_preencher.remove("resposta_NPS")
```

```python
colunas_para_preencher.remove("metadata_codcliente")
```

```python
colunas_para_preencher
```

```python
coluna_base = "resposta_NPS"

for col in colunas_para_preencher:
    df[col] = df[col].fillna(df[coluna_base])
```

```python
df_agregado = df.groupby('metadata_codcliente').agg('mean').reset_index()
```

```python
df_agregado = df_agregado.rename(columns={'metadata_codcliente': 'CD_CLIENTE'})
```

```python
df_agregado
```

```python
df_agregado.isnull().sum()/len(df_agregado)
```

```python
df_clientes_clusterizados = pd.read_csv('../dados/dados_tratados_v0/clientes_clusterizados_v1.csv', sep=';')
```

```python
df_clientes_clusterizados
```

```python
df_clientes_clusterizados = pd.merge(
    df_clientes_clusterizados, 
    df_agregado, 
    on="CD_CLIENTE",   # chave comum
    how="left"        # tipo de junção
)
```

```python
df_clientes_clusterizados
```

```python
# Arredondando valores float para 2 casas decimais

cols_float = df_clientes_clusterizados.select_dtypes(include='float').columns
df_clientes_clusterizados[cols_float] = df_clientes_clusterizados[cols_float].round(2)
```

```python
df_clientes_clusterizados.to_csv('../dados/dataset_final_v0/clientes_clusterizados_v2_com_nps.csv', index=False, sep=';', decimal = ',')
```

```python
df_agregado.to_csv('../dados/dados_tratados_v0/nps_relacional_tratado.csv', index=False, sep=';')
```
